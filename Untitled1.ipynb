```python
import numpy as np
import pandas as pd
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
df = pd.DataFrame(exam_data, index=labels)
df
```

```python
df.head(3)
```

```python
df.drop(["d","h"])
```

```python
df = df.dropna()
```

```python
df[["name","score"]]
```

```python
new_line = {"name": "Suresh", "score": 15.5, "attempts": 1, "qualify": "yes"}
new_line = pd.DataFrame({"k": new_line}).T
df = pd.concat([df, new_line])
```

```python
df.drop("attempts",axis=1)
```

```python
df['Success'] = df['score'].apply(lambda x: 1 if x > 10 else 0) #Méthode 1
# df['Success'] = (df["score"] > 10).astype(int) #Méthode 2
df
```

```python
df.to_csv("my_data.csv")
```
